### Question 1: Convert the datatype of 'Date' column to Date-Time format.

```python
df.dtypes
```

```python
df.Date = pd.to_datetime(df.Date)
```

```python
df.dtypes
```

### Question 2: Add a new column 'Year' in the dataframe, which contains year only.

```python
df['Year'] = df.Date.dt.year
```

```python
df.head()
```

### Question 3: Add a new column 'Month' as second column in the dataframe, which contains Month only.

```python
df.insert(2, 'Month', df.Date.dt.month)
```

```python
df.head()
```

### Question 4: Remove the columns 'Year' and 'Month' from the dataframe.

```python
df.drop(['Month', 'Year'], axis=1,  inplace = True)
```

```python
df.head()
```

### Question 6: What is the maximum and minimum 'Price' per year in Ibeju-Lekki?

```python
df1=df[df.Area=='Ibeju-Lekki']
```

```python
df1
```

```python
df1.groupby('Date').Price.max() #Maximum price
```

```python
df1.groupby('Date').Price.min()  #Minimum price
```

```python
df1.groupby('Date').Price.mean()  #Mean price
```

### Question 7: What is the maximum and minimum 'Price' recorded by area?

```python
df.groupby('Area').Price.max().sort_values(ascending = True) #Maximum price by Area in Ascending Order
```

```python
df.groupby('Area').Price.min().sort_values(ascending = True) #Minimum price by Area in Ascending Order
```
